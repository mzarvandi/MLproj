## 1. Find & Remove Highly Correlated Features

```python
%%script false --no-raise-error  # Disable this method

import numpy as np

new_dfs = []
for idx in range(len(dfs)):
    fts = dfs[idx].shape[1]
    numeric = dfs[idx].drop(['Age','Status', 'Gender'], axis=1)
    correlation = numeric.corr()
    numeric_columns = numeric.columns
    high_corr = [ ]

    for c1 in numeric_columns:
      for c2 in numeric_columns:
        if (c1 != c2) and (c1 not in high_corr) and (correlation[c1][c2] > 0.9):
              high_corr.append(c2)
    high_corr = np.unique(np.array(high_corr))
    new_dfs.append(dfs[idx].drop(high_corr, axis=1))
    print(f"{len(high_corr)} of {fts} features found as correlated and removed in df_f{idx+1}")
```

## 2. Find Best Number of Features for Each Label

```python
import numpy as np

from sklearn.feature_selection import SelectFromModel
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier

trains_x = {}
trains_y = []
for l, label in enumerate(['Age', 'Status', 'Gender']):
    new_train = []
    for idx in range(len(dfs)):
        data = dfs[idx].values
        x = data[:, 3:]
        trains_y.append(data[:, :3])
        
        y = trains_y[idx][:, l]
        model = RandomForestRegressor(random_state=0, n_jobs=-1)
        
        if label != 'Age':
            le = LabelEncoder()
            le.fit(np.unique(y))
            y = le.transform(y)
            model = RandomForestClassifier(random_state=0, n_jobs=-1)
        
        selector = SelectFromModel(estimator=model).fit(x, y)
        new_train.append(selector.transform(x))
        trains_x[label] = new_train
        print(f"New shape of df_f{idx+1} for {label} is {trains_x[label][idx].shape}")
    print("="*30)
```

## 4. Sequential Feature Selector Method

```python
%%script false --no-raise-error  # Disable this method



from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier

sfs_trains_x = {}
sfs_trains_y = []
for l, label in enumerate(['Age', 'Status', 'Gender']):
    new_train = []
    for idx in range(len(dfs)):
        data = dfs[idx].values
        x = data[:, 3:]
        sfs_trains_y.append(data[:, :3])
        
        y = sfs_trains_y[idx][:, l]
        model = RandomForestRegressor(random_state=0, n_jobs=-1)
        
        if label != 'Age':
            le = LabelEncoder()
            le.fit(np.unique(y))
            y = le.transform(y)
            model = RandomForestClassifier(random_state=0, n_jobs=-1)
        
        sfs = SequentialFeatureSelector(estimator=model,direction='forward', n_jobs=-1).fit(x, y)
        new_train.append(sfs.transform(x))
        sfs_trains_x[label] = new_train
        print(f"New shape of df_f{idx+1} for {label} is {sfs_trains_x[label][idx].shape}")
    print("="*30)
```

```python
from imblearn.over_sampling import SMOTENC
from imblearn.over_sampling import RandomOverSampler 


def oversampling(_X, _Y):        
    oversample = SMOTENC(categorical_features=[0], random_state = 100)
    return oversample.fit_resample(_X, _Y)

def oversampling_random_over_sampler(_X, _Y):
    oversample = RandomOverSampler(random_state=100)
    return oversample.fit_resample(_X, _Y)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
import imblearn

_ = {}
for label, dfs_list in reduced_dfs.items():
    for idx, reduced_df  in dfs_list.items():                
        _X_Train, _X_Test, _Y_label_Train, _Y_label_Test = train_test_split(
            reduced_df.drop([label, ], axis=1), 
            reduced_df[label], test_size=0.3, train_size=0.7,
            random_state=None, shuffle=True, stratify=None
        )

        if label=='Age':
            _X_Train, _Y_label_Train = oversampling_random_over_sampler(_X_Train, _Y_label_Train)
        else:
            _X_Train, _Y_label_Train = oversampling(_X_Train, _Y_label_Train)
        _train = _X_Train.merge(_Y_label_Train, left_index=True, right_index=True)
        _test = _X_Test.merge(_Y_label_Test, left_index=True, right_index=True)

        if _.get(label, None) is None:           
            _[label] = {}

        _[label][idx] = {
            'test': _test,
            'train': _train
        }
        
        print(f'We have {reduced_df.shape} for df{idx+1} with label {label} --> {_train.shape}')

        
    
upsampled_dfs = _
```

# Clustering

```python
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
# from yellowbrick.cluster import KElbowVisualizer
from scipy.cluster.hierarchy import dendrogram
from sklearn.cluster import AgglomerativeClustering
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.preprocessing import OneHotEncoder
```

```python
def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack([model.children_, model.distances_,
                                      counts]).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)
```

```python
def clustering(X, label):
    if label == 'Status':
        X[label] = X[label].map({'happy': 0, 'natural':1, 'sad':2})
    elif label == 'Gender':
        X[label] = X[label].map({'male': 0, 'female':1})

    # setting distance_threshold=0 ensures we compute the full tree.
    model = AgglomerativeClustering(distance_threshold=0, n_clusters=None, compute_distances=True)
    model = model.fit(X)
    
    plt.figure(figsize=(15, 8))
    plt.title('Hierarchical Clustering Dendrogram')
    # plot the top three levels of the dendrogram
    plot_dendrogram(model, truncate_mode='level', p=4)
    plt.xlabel("Number of points in node (or index of point if no parenthesis).")
    plt.show()
```

```python
def findK(X, label):
    if label == 'Status':
        X[label] = X[label].map({'happy': 0, 'natural':1, 'sad':2})
    elif label == 'Gender':
        X[label] = X[label].map({'male': 0, 'female':1})
        
    kmodel = KMeans()
    visualizer = KElbowVisualizer(kmodel, k=(1, 20))

    visualizer.fit(X)        # Fit the data to the visualizer
    visualizer.show()        # Finalize and render the figure
```

```python
def fit_kmeans(X, label, k):
    if label == 'Status':
        X[label] = X[label].map({'happy': 0, 'natural':1, 'sad':2})
    elif label == 'Gender':
        X[label] = X[label].map({'male': 0, 'female':1})
        
    kmeans = KMeans(n_clusters = k, random_state = 0).fit(X)
    x_pca = PCA(2).fit_transform(X)
    x_tsne = TSNE().fit_transform(X)
    
    
    return x_pca, x_tsne, kmeans
```
